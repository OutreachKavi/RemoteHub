```python
import pickle as pkl
```

```python
pkfl = pkl.load(open('/content/model (1).pkl', 'rb'))
```

```python
age = int(input("Enter AGE: "))
Smoking = int(input("Does the person smokes "))
YF = int(input("Does the person has yellow fingers: "))
anxiety = int(input("Does the person feels anxiety: "))
CD = int(input("Does the person has Chronic Disease?: "))
AC = int(input("Does the person consumes alcohol?: "))
cough = int(input("Does the person has frequent Cough?: "))
SB = int(input("Does the person has Shortness of Breath?: "))
CP = int(input("Does the person has Chest Pain?: "))
LUNG_CANCER = pkfl.predict([[age,Smoking,YF,anxiety,CD,AC,cough,SB,CP]])
print("Lung_Cancer : ", LUNG_CANCER[0])
```
