```python
import pickle as pkl
```

```python
pkfl = pkl.load(open('/content/model (1).pkl', 'rb'))
```

```python
Diagnosis = int(input("Enter the diagnosis "))
FVC = int(input("Enter the FVC "))
FEV1 = int(input("Enter the FEV1 Value "))
Performance = int(input("Enter the performance value "))
Pain = int(input("Does the patiend have pain "))
Haemoptysis= int(input("Does the person feel haemoptysis: "))
Dyspnoea = int(input("Does the person feel Dyspnoea: "))
Cough = int(input("Does the person has cough "))
Weakness = int(input("Does the person feel weak "))
Tumor_Size = int(input("What is the tumor size of the paitent "))
Diabetes_Mellitus  = int(input("does the patient has diabete mellitus "))
MI_6mo  = int(input("Does the person as MI_6mo "))
PAD  = int(input("enter the value for PAD "))
Smoking = int(input("Does the person smokes "))
Asthma = int(input("Does the person has asthma "))
Age = int(input("enter the age of the patient "))
Death_1yr = pkfl.predict([[Diagnosis ,FVC ,FEV1,Performance,Pain, Haemoptysis,Dyspnoea , Cough,Weakness,Tumor_Size ,Diabetes_Mellitus ,MI_6mo ,PAD, Smoking,Asthma,Age]])
print("Death_1yr: ", Death_1yr[0])
```
